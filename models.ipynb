```python
"""
Reusable model pipelines for the Credit Risk project.
"""

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

def logistic_regression_pipeline(preprocess, class_weight=None, C=1.0):
    return Pipeline([
        ("preprocess", preprocess),
        ("model", LogisticRegression(
            max_iter=1000,
            class_weight=class_weight,
            C=C,
            solver="lbfgs"
        ))
    ])

```

```python
def random_forest_pipeline(preprocess, class_weight="balanced"):
    return Pipeline([
        ("preprocess", preprocess),
        ("model", RandomForestClassifier(
            n_estimators=200,
            class_weight=class_weight,
            random_state=42,
            n_jobs=-1
        ))
    ])
```
